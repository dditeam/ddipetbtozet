```python
import jpype
import pandas as pd
import numpy as np
import re

libjvmpath = r"C:\Program Files\Java\jdk-18\bin\server\jvm.dll" 
jar = "zemberek-full.jar" 

if not jpype.isJVMStarted():
    jpype.startJVM(jvmpath=libjvmpath, classpath=[jar])
from jpype import JClass, JString, java

df = pd.read_csv("final.csv")
df['bulgular1'] = df['bulgular1'].astype("string")
df['bulgular2'] = df['bulgular2'].astype("string")
df.fillna('', inplace=True)
df['bulgular'] = df.bulgular1.str.cat(df.bulgular2)

datas = pd.read_csv(r"C:\Users\ersam\OneDrive\Desktop\raporlar.csv")

```

```python
for i in range(0,datas['bulgular'].size):
    datas['bulgular'].loc[datas.index[i]] =re.sub(r'(?<=[.,])(?=[^\s])+(?![0-9])', r' ',str(datas['bulgular'].loc[datas.index[i]])) 
    datas['bulgular'].loc[datas.index[i]] =re.sub('\n', ' ',str(datas['bulgular'].loc[datas.index[i]]))
    datas['bulgular'].loc[datas.index[i]]=re.sub(" '","'",datas['bulgular'].loc[datas.index[i]])
    datas['bulgular'].loc[datas.index[i]]=re.sub("' ","'",datas['bulgular'].loc[datas.index[i]])
    datas['bulgular'].loc[datas.index[i]]=re.sub(" +"," ",datas['bulgular'].loc[datas.index[i]])
    datas['bulgular'].loc[datas.index[i]]=datas['bulgular'].loc[datas.index[i]].rstrip()
bulgular1 = datas.bulgular

for i in range(0,df['bulgular'].size):
    df['bulgular'].loc[df.index[i]] =re.sub(r'(?<=[.,])(?=[^\s])+(?![0-9])', r' ',str(df['bulgular'].loc[df.index[i]])) 
    df['bulgular'].loc[df.index[i]] =re.sub('\n', ' ',str(df['bulgular'].loc[df.index[i]]))
    df['bulgular'].loc[df.index[i]]=re.sub(" '","'",df['bulgular'].loc[df.index[i]])
    df['bulgular'].loc[df.index[i]]=re.sub("' ","'",df['bulgular'].loc[df.index[i]])
    df['bulgular'].loc[df.index[i]]=re.sub(" +"," ",df['bulgular'].loc[df.index[i]])
    df['bulgular'].loc[df.index[i]]=df['bulgular'].loc[df.index[i]].rstrip()
bulgular2 = df.bulgular
```

```python
WordAnalysis: JClass = JClass('zemberek.morphology.analysis.WordAnalysis')
TurkishMorphology:JClass = JClass('zemberek.morphology.TurkishMorphology')
SingleAnalysis:JClass = JClass('zemberek.morphology.analysis.SingleAnalysis')
morphology: TurkishMorphology = TurkishMorphology.createWithDefaults()
TurkishTokenizer: JClass = JClass('zemberek.tokenization.TurkishTokenizer')
TokenIterator: JClass = JClass('zemberek.tokenization.TurkishTokenizer.TokenIterator')
Token: JClass = JClass('zemberek.tokenization.Token')
tokenizer: TurkishTokenizer = TurkishTokenizer.DEFAULT
word_list = []
number = len(bulgular1) + len(bulgular2)
docs = [""] * number

print(len(docs))

for j in range(len(bulgular1)):  
    token_iterator: TokenIterator = tokenizer.getTokenIterator(JString(bulgular1[j]))
    for i in token_iterator:
        word = i.content
        results: WordAnalysis  = morphology.analyze(JString(word))
        result = results.getAnalysisResults()
        state = result.isEmpty()
        if state == False:
            a = (result.get(0)).getLemmas()
            docs[j] = docs[j] + " "+ str(a[0])
        else:
            docs[j] = docs[j] +" " + str(word)
    
for j in range(len(bulgular2)):  
    token_iterator: TokenIterator = tokenizer.getTokenIterator(JString(bulgular2[j]))
    for i in token_iterator:
        word = i.content
        results: WordAnalysis  = morphology.analyze(JString(word))
        result = results.getAnalysisResults()
        state = result.isEmpty()
        if state == False:
            a = (result.get(0)).getLemmas()
            docs[j + len(bulgular1)] = docs[j + len(bulgular1)] + " "+ str(a[0])
        else:
            docs[j + len(bulgular1)] = docs[j + len(bulgular1)] +" " + str(word)
```
